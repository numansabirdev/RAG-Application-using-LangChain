```python
from langchain_community.llms import Ollama

llm = Ollama(model="llama3",temperature=0)
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('constitution of pakistan.pdf')
docs = loader.load_and_split()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(docs)
```

```python
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()

vectorstore = FAISS.from_documents(chunks, embeddings)

retriever = vectorstore.as_retriever()
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

contextualize_q_system_prompt = (
    "Given a chat history and the latest user question "
    "which might reference context in the chat history, "
    "formulate a standalone question which can be understood "
    "without the chat history. Do NOT answer the question, "
    "just reformulate it if needed and otherwise return it as is."
)
contextualize_q_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", contextualize_q_system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)
```

```python
from langchain.chains import create_history_aware_retriever

history_aware_retriever = create_history_aware_retriever(
    llm, retriever, contextualize_q_prompt
)
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

template =  (
    "You are a helpful and knowledgeable ChatBot. Answer the question based"
    "on the context below, and respond naturally with concise answers, as if you were a"
    "human. Avoid phrases like Based on the information provided or"
    "The text you provided or Based on the provided metadata Just provide the"
    "answer directly. If you don't know the answer, just say that you don't know,"
    "don't try to make up an answer"
    "\n\n"
    "{context}"
    "\n\n"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", template),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)
```

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain

question_answer_chain = create_stuff_documents_chain(llm, prompt)

rag_chain = create_retrieval_chain(history_aware_retriever, question_answer_chain)
```

```python
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory


store = {}


def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]


chain = RunnableWithMessageHistory(
    rag_chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history",
    output_messages_key="answer",
)
```

```python
import time

time_1 = time.time()

response = chain.invoke(
    {"input": "What is this document related"},
    config={
        "configurable": {"session_id": "asd"}
    }
)["answer"]

time_2 = time.time()

elapsed_time_seconds = time_2 - time_1

elapsed_time_minutes = elapsed_time_seconds / 60

print(response)
print(f"Total time: {elapsed_time_minutes:.2f} minutes")
```

```python
session_id = "asd1"

while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit', 'quit', 'stop', 'bye']:
        break

    response = chain.invoke(
        {"input": user_input},
        config={
            "configurable": {"session_id": session_id}
        }
    )["answer"]

    print(f"Bot: {response}")
```

```python
from langchain_core.messages import AIMessage

for message in store["asd1"].messages:
    if isinstance(message, AIMessage):
        prefix = "Bot"
    else:
        prefix = "User"

    print(f"{prefix}: {message.content}\n")
```
